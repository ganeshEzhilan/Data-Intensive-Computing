### Problem 3

```r
fn<-boxplot(sales,col=c("orange","green"))$stats
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

#### Interpretation: The thrid quartile(50%-75%) of the 'Sales 1' has higher variance. More than 75% of the 'Sales2' values are greater than the median value of 'Sales1'.
